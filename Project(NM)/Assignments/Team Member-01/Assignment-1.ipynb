```python
sns.rugplot(x='lot_area', data=df, height=.03, color='blue')
sns.histplot(x='lot_area', data=df, kde=True,color='red')
```

MULTIVARIATE ANALYSIS

```python
sns.pairplot(df,vars=['lot_area','living_area','number_of_bedrooms','number_of_floors','number_of_views'],hue='number_of_bathrooms',diag_kind='kde')
plt.suptitle('Multivariate Analysis')
plt.show()
```
