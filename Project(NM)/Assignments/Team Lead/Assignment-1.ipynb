##MULTIVARIATE ANALYSIS

```python
sns.pairplot(df,vars=['living_area','lot_area','number_of_floors','number_of_bedrooms','number_of_bathrooms'],hue='number_of_views',diag_kind='kde')
plt.suptitle('Multivariate Analysis')
plt.show()
```
