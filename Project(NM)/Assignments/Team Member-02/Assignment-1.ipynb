```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/House Price India.csv")
df.head()
```

**UNIVARIATE ANALYSIS**

```python
sns.histplot(x='Built Year', data=df);
```

```python
df1 = df.head(40)
```

```python
sns.kdeplot(x='Distance from the airport', data=df1)
```

```python
x = df1['living_area'].value_counts()
plt.pie(x.values,
        labels=x.index,
        autopct='%1.1f%%')
plt.show()
```

```python
sns.histplot(x='living_area', data=df1, kde=True)
```

```python
sns.rugplot(x='lot_area', data=df, height=.03, color='tomato')
sns.histplot(x='lot_area', data=df, kde=True,color='green')
```

```python
sns.violinplot(x=df["lot_area"],color='grey')
```
